```python
divisao = filmes['genres'].str.split('|')
generos = []
for row in divisao:
    for elem in row:
        generos.append(elem)
generos = set(generos)
generos
```

```python
for i in generos:
    filmes[i] = 0
```

```python
divisao
```

```python
counter = 0
for row in divisao:
    for i in row:
        filmes.loc[counter, i] = 1
    counter += 1
```
